```python
# python example
# author: Gannon Ellert

import pandas as pd
import numpy as np

filepath = 'data_test_1.xlsx'
people_data = pd.read_excel(filepath, sheet_name='people_data')

#   ======  Continue with your code  =========

revenue_data = pd.read_excel(filepath, sheet_name='revenue_data')

# I used Jupyter notebooks to work on and test this code. Below are the steps I took:



# Getting familiar with the data:

print("people data: ", people_data.shape)
print("revenue data: ", revenue_data.shape)

# people_data.dtypes
# people_data.isnull().sum()
# people_data.nunique()
# people_data.sample(10)

# revenue_data.dtypes
# revenue_data.sample(10) 
# revenue_data.isnull().sum()
# revenue_data.nunique()

print()


# Merge and review Revenue and People data frames:

df = revenue_data.merge(people_data, on='employee_id', how='left')
print("merge shape: ", df.shape)
print("min date: ", df['customer_close_date'].min())
print("max date: ", df['customer_close_date'].max())
print()
# df.nunique()



# Create column to specify month of sale:

df['month'] = pd.DatetimeIndex(df['customer_close_date']).month
print("new shape with month: ", df.shape)
print()
print("sample:")
print(df.sample(5))
print()
print()



# Omit Eliana Bernard and subsequent data:

df = df[df.manager_name != "Eliana Bernard"]
print("data without Eliana Bernard: ", df.shape)
print()
# df.nunique()



# Pivot dataframe:

df_pivot = pd.pivot_table(data=df, index='team', columns='month', values='revenue_amount', aggfunc='sum')
print("pivoted data frame:")
print(df_pivot)
print()



# Flatten pivot table into new dataframe:

df_flattened = pd.DataFrame(df_pivot.to_records())



# Replace "NaN" values with 0, add an addition column for total_revenue:

df_flattened = df_flattened.fillna(0)
df_flattened['total_revenue'] = df_flattened["1"] + df_flattened["2"] + df_flattened["3"]


# Rename columns:

df_flattened = df_flattened.rename(columns = {"1": "JAN",
                                             "2": "FEB",
                                             "3": "MAR"})


# Final product:
print()
print("fnal table:")
print(df_flattened)



# Write final dataframe to CSV file:

df_flattened.to_csv('twilio_final.csv', index=False)
```
