```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
lbdata = pd.read_csv('DataSets/weight-height.csv')

lbdata.shape
```

```python
lbdata.head()
```

```python
#frequency distribution
# curve line = probability distribution 

plt.figure(figsize=(12,8))

sns.distplot(lbdata['Weight'], color='purple')
plt.title('weights')
plt.ylabel('count')
```

```python
plt.figure(figsize=(12,8))

sns.distplot(lbdata['Weight'], color='purple', bins=70)
plt.title('weights')
plt.ylabel('density')
```

```python
plt.figure(figsize=(12,8))

sns.distplot(lbdata['Weight'], color='purple', bins=70, hist=False)
plt.title('weights')
plt.ylabel('density')
```

```python
plt.figure(figsize=(12,8))

sns.distplot(lbdata['Weight'], color='purple', bins=70, kde=False, rug=True)
plt.title('weights')
plt.ylabel('density')
```

```python
plt.figure(figsize=(12,8))

sns.set(color_codes=True)
sns.kdeplot(lbdata['Weight'], shade=True, color='green')
#sns.distplot(lbdata['Weight'], color='purple', bins=70, hist=False)
plt.title('weights')
plt.ylabel('density')
```

```python
car = pd.read_csv('DataSets/car-data.csv')
car.head()
```

```python
car.isna().sum()
```

```python
car.replace('.', np.nan, inplace=True)
car.replace('?', np.nan, inplace=True)
```

```python
car = car.dropna()
```

```python
car.shape
```

```python
plt.figure(figsize=(10,8))

sns.scatterplot(x='hp', y='mpg', data=car)

plt.xlabel('hp')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,8))

chart = sns.scatterplot(x='hp', y='mpg', data=car, hue='cyl')

chart.set_xticklabels(chart.get_xticklabels(), rotation=45)


plt.xlabel('hp')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,8))

chart = sns.scatterplot(x='hp', y='mpg', data=car, hue='cyl', size='lb', sizes=(30, 300), hue_norm=(0,9))

chart.set_xticklabels(chart.get_xticklabels(), rotation=45)


plt.xlabel('hp')
plt.ylabel('mpg')
plt.show()
```

```python
car.hp = car.hp.astype(int)
```
