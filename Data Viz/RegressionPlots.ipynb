```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
car = pd.read_csv('DataSets/car-data.csv')

car.sample(5)
```

```python
car.replace('.', np.nan, inplace=True)
car.replace('?', np.nan, inplace=True)

car = car.dropna()
car.head()
```

```python
car.hp = car.hp.astype(int)
car.mpg = car.mpg.astype(int)
car.acceleration = car.acceleration.astype(int)
```

```python
sns.lmplot(x='hp', y='mpg', data = car)
```

```python
sns.lmplot(x='hp', y='mpg', hue='origin', aspect=1.5, data=car)
```

```python
sns.lmplot(x='hp', y='mpg', hue='origin', palette='Set2', aspect=1.5, data=car)
```

```python
# regression through categorical data

sns.lmplot(x='cyl', y='mpg', aspect=1.5, data=car)
```

```python
sns.lmplot(x='lb', y='mpg', data=car, col='origin')
```

```python
sns.pairplot(car, height=3, vars=['mpg', 'acceleration', 'hp'], kind='reg')
```

```python
car.to_csv('car.csv')
```
