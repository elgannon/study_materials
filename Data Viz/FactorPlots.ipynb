```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
car = pd.read_csv('DataSets/car.csv')
car.sample(5)
```

```python
f, ax = plt.subplots(figsize=(12,8))

sns.violinplot(x='year', y='mpg', data=car)
```

```python
f, ax = plt.subplots(figsize=(12,8))

sns.barplot(x='cyl', y='mpg', data=car)
```

```python
f, ax = plt.subplots(figsize=(12,8))

sns.pointplot(x='cyl', y='mpg', data=car)
```

```python
sns.catplot(x='cyl', data=car, height=5, aspect=2, kind="count")

# (frequncy chart)
```
