```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
car = pd.read_csv('DataSets/car-data.csv')

car.head()
```

```python
car.replace('.', np.nan, inplace=True)
car.replace('?', np.nan, inplace=True)

car = car.dropna()
car.head()
```

```python
car.shape
```

```python
car_subset = car[['acceleration', 'hp', 'lb', 'mpg']]
```

```python
car_subset.dtypes
```

```python
car_subset.hp = car.hp.astype(int)
car_subset.mpg = car.mpg.astype(int)
car_subset.acceleration = car.acceleration.astype(int)
```

```python
car_subset.dtypes
```

```python
sns.pairplot(car_subset)
```

```python
sns.pairplot(car, 
            vars=['displacement','hp','lb'],
            hue='cyl')
```

```python
# compare pairs of vars

corrmat = car.corr()
f, ax = plt.subplots(figsize=(12,8))

sns.heatmap(corrmat, vmax=.8, square=True, annot=True,
            fmt='.2f', cmap = 'viridis')

plt.show()
```
