```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
car = pd.read_csv('DataSets/car.csv')
```

```python
car.sample(5)
```

```python
plt.figure(figsize=(12,8))

sns.stripplot(car['mpg'])
```

```python
plt.figure(figsize=(12,8))

sns.stripplot(x='cyl', y='mpg', data=car, jitter=True)

#jitter prevents overlap
```

```python
plt.figure(figsize=(12,8))

sns.stripplot(x='cyl', y='mpg', hue='origin', data=car, jitter=True)
```

```python
plt.figure(figsize=(12,8))

sns.swarmplot(x='cyl', y='mpg', data=car)
```

```python
#swarm plots will not scale well with larger datasets
```

```python
plt.figure(figsize=(12,8))

sns.swarmplot(x='cyl', y='mpg', hue='origin', dodge=True, data=car)

#dodge separates data by cat specified
```
